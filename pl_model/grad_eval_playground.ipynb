```python
from transformers import BertTokenizer, CONFIG_MAPPING, BertModel
from datasets import load_from_disk, load_dataset
```

```python
config = CONFIG_MAPPING['bert'].from_pretrained('bert-base-uncased')
```

```python
bert = BertModel(config,add_pooling_layer=False).from_pretrained('bert-base-uncased')
```

```python
bert.training
```

```python
len(bert.encoder.layer)
```

```python
bert.encoder.training
```

```python
for name,param in bert.named_parameters():
    print(f"{name} : {param.requires_grad}")
```

```python
num_trainable_layers = 3
```

```python
bert.train()
```

```python
for param in bert.encoder.layer[-num_trainable_layers:].parameters():
    param.requires_grad=False
```

```python
for layer in bert.encoder.layer[:bert.config.num_hidden_layers-12]:
    print(layer)
    layer.eval()
    for param in layer.parameters():
        param.requires_grad=False
```

```python
for layer in bert.encoder.layer:
    print(layer.training)
```
