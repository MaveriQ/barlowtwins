```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def process_data(filename):
    model_lars = pd.read_csv(filename, usecols=['epoch','step','STSBenchmark','SICKRelatedness'])

    if model_lars.step.dtype=='object':
        model_lars.step = model_lars.step.apply(lambda x: x.replace('.ckpt','')).astype('int32')
    else:
        model_lars.step = model_lars.step.astype('int32')

    model_lars = model_lars.reindex(model_lars.sort_values(by=['epoch','step']).index,axis=0).reset_index(drop=True)

    model_lars = model_lars.reindex(['epoch','step','STSBenchmark','SICKRelatedness'],axis=1)
    
    fig, ax = plt.subplots(1,2)
    fig.set_size_inches(20, 10)
    ax[0].plot(model_lars.step,model_lars.SICKRelatedness)
    ax[1].plot(model_lars.step,model_lars.STSBenchmark);
    ax[0].legend(['SICKRelatedness'])
    ax[1].legend(['STSBenchmark']);
    
    return model_lars
```
