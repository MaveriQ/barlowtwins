```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def process_data(filename):
    model_lars = pd.read_csv(filename, usecols=['epoch','step','STSBenchmark','SICKRelatedness'])

    if model_lars.step.dtype=='object':
        model_lars.step = model_lars.step.apply(lambda x: x.replace('.ckpt','')).astype('int32')
    else:
        model_lars.step = model_lars.step.astype('int32')

    model_lars = model_lars.reindex(model_lars.sort_values(by=['epoch','step']).index,axis=0).reset_index(drop=True)

    model_lars = model_lars.reindex(['epoch','step','STSBenchmark','SICKRelatedness'],axis=1)
    return model_lars
```

```python
model_lars = process_data('bert_frozen_bs1024_dropout0.1_lr1e-3_lambd1e-3_mlm0_gpus4_mean_all_fp16_l_a_r_s.csv')
```

```python
model_768 = process_data('bert_frozen_bs1024_dropout0.1_lr1e-3_lambd1e-3_mlm0_gpus4_mean_all_fp16_768-768-768.csv')
```

```python
model_2048 = process_data('bert_frozen_bs1024_dropout0.1_lr1e-3_lambd1e-3_mlm0_gpus4_mean_all_fp16_2048-2048-2048.csv')
```

```python
model_768.STSBenchmark.plot()
model_2048.STSBenchmark.plot()
plt.legend(['lars','768','2048']);
```

```python
model_lars.SICKRelatedness.plot()
model_768.SICKRelatedness.plot()
model_2048.SICKRelatedness.plot();
plt.legend(['lars','768','2048']);
```
